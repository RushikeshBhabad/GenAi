```python
from llama_index.core import SummaryIndex, VectorStoreIndex
```

```python
summary_index = SummaryIndex(nodes) # try to build summary
vector_index = VectorStoreIndex(nodes) # used for rag use cases when user asks document specific questions
```

```python
summary_query_engine = summary_index.as_query_engine(
    response_mode="tree_summarize",
    use_async=True
)

vector_query_engine = vector_index.as_query_engine()
```

```python
from llama_index.core.tools import QueryEngineTool
```

```python
summary_tool = QueryEngineTool.from_defaults(
    query_engine=summary_query_engine,
    description = (
        "Useful for summarization related to the given document"
    )
)

vector_tool = QueryEngineTool.from_defaults(
    query_engine=vector_query_engine,
    description = (
        "Useful for retrieving specific context from given context based on the given question"
    )
)
```

```python
from llama_index.core.query_engine.router_query_engine import RouterQueryEngine
from llama_index.core.selectors import LLMSingleSelector
```

```python
query_engine = RouterQueryEngine(
    selector = LLMSingleSelector.from_defaults(),
    query_engine_tools=[
        summary_tool,
        vector_tool
    ],
    verbose=True
)
```

```python
response = query_engine.query("Summarize the given document")
print(response)
```

```python
response = query_engine.query("What is the black box and white box testing ?")
print(response)
```
